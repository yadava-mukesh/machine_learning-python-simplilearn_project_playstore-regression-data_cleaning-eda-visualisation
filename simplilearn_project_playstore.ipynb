# 4 .Variables seem to have incorrect type and inconsistent formatting. You need to fix them:
 ***4.1 Size column has sizes in Kb as well as Mb. To analyze, you’ll need to convert these to numeric.***
- ***4.1.1*** Extract the numeric value from the column
- ***4.1.2*** Multiply the value by 1,000, if size is mentioned in Mb

```python
def convert_size(size):
    if 'M' in size:
        x=size[:-1]
        x=float(x)*1000
        return(x)
    elif 'k'==size[:-1]:
        x=size[:-1]
        x=float(x)
        return(x)
    else:
        return None
```

```python
cleaned_data["Size"]=cleaned_data["Size"].map(convert_size)
```

```python
cleaned_data.Size.describe()
```

```python
cleaned_data.Size.fillna(method='ffill', inplace=True)
```

***4.3.Installs field is currently stored as string and has values like 1,000,000+.***

```python
cleaned_data['Installs'].value_counts()
```

# 8. Data preprocessing

For the steps below, create a copy of the dataframe to make all the edits. Name it inp1.

***8.1*** Reviews and Install have some values that are still relatively very high. Before building a linear regression model, 
    you need to reduce the skew. Apply log transformation (np.log1p) to Reviews and Installs.

```python
cleaned_data.columns
```

```python
cleaned_data['Reviews'].skew() #right skewed
```

```python
cleaned_data['Installs'].skew() #right skewed
```

```python
cleaned_data.shape
```

```python
inp1=cleaned_data.copy()
```

```python
inp1['log1p_Reviews']=np.log1p(inp1['Reviews'])
inp1['log1p_Reviews'].skew()
```

Review is now Normal distributted.

```python
inp1['log1p_Installs']=np.log1p(inp1['Installs'])
inp1['log1p_Installs'].skew()
```

Installs is now Normal distributted.

***8.2*** Drop columns App, Last Updated, Current Ver, and Android Ver. These variables are not useful for our task.

```python
inp1.head(3)
```

```python
inp1.columns
```

```python
inp1.drop(['App','Reviews','Installs','Last Updated', 'Current Ver',
       'Android Ver'], axis=1, inplace=True)
```

***8.3*** Get dummy columns for Category, Genres, and Content Rating. This needs to be done as the models do not understand categorical data, and all data 
should be numeric. Dummy encoding is one way to convert character fields to numeric. Name of dataframe should be inp2.

```python
inp1.shape
```

```python
inp2 = pd.get_dummies(inp1, columns=['Category','Type','Genres','Content Rating'], dtype=int, drop_first=True)
inp2.head(3)
```

```python
inp2.shape
```

```python
X=inp2.drop('Rating', axis=1)
y=inp2['Rating']
```

```python
X.shape
```

```python
y.shape
```

# 9. Train test split  and apply 70-30 split
# 10. Separate the dataframes into X_train, y_train, X_test, and y_test.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=100)
```

# 11 . Model building

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#lr.coef_
lr.intercept_
```

```python
ypred_train=lr.predict(X_train)
print('ytrain predicted is: ', ypred_train)
from sklearn.metrics import r2_score
print('R2 score is: ', r2_score(y_train, ypred_train))
```

```python
ypred_test=lr.predict(X_test)
print('ytest predicted is: ', ypred_test)
from sklearn.metrics import r2_score
print('R2 score is: ', r2_score(y_test, ypred_test))
```
